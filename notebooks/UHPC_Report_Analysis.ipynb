```python
clean_df = pd.read_csv("../data/UHPC_Cleaned.csv")

print(clean_df.shape)
```

```python
print(clean_df.columns.tolist())
```

```python
selected_columns = [
    'Cement Amount  (kg/m³)',
    'Cement Grade (Mpa)',
    'Silica Fume (kg/m³)',
    'Flayash Amount   (kg/m³)',
    'Limestone Powder  (kg/m3)',
    'Quartzpowder (kg/m3)',
    'Glass powder (kg/m3)',
    'Rice husk ash (kg/m3)',
    'Metakaolin (kg/m³)',
    'GGBFS  (kg/m³)',
    'Slag Amount   (kg/m³)',
    'Nano-CaCO3 (kg/m3)',
    'Nano-Al2O3 (kg/m3)',
    'Nano-TiO2 (kg/m3)',
    'Nano Silica (kg/m3)',
    'Filler (kg/m³)',
    'Amount  (kg/m³)',          # Sand
    'Amount / Quantity of Fiber',
    'Amount  (kg/m³).2',        # Superplasticizer
    '28-Day'                    # Target (Compressive Strength)
]
```

```python
table_3_1 = (
    clean_df[selected_columns]
    .describe()
    .T[['min', 'max', 'mean', 'std']]
    .round(2)
)

table_3_1.columns = [
    'Minimum',
    'Maximum',
    'Mean',
    'Standard Deviation'
]

table_3_1
```

```python
table_3_1.to_excel(
    "Tables/Table_3_1_Descriptive_Statistics.xlsx"
)

print("Table 3.1 saved successfully!")
```

```python
# Rename variables for the report

report_df = clean_df[selected_columns].rename(columns={
    'Cement Amount  (kg/m³)': 'Cement (kg/m³)',
    'Cement Grade (Mpa)': 'Cement Grade (MPa)',
    'Silica Fume (kg/m³)': 'Silica Fume (kg/m³)',
    'Flayash Amount   (kg/m³)': 'Fly Ash (kg/m³)',
    'Limestone Powder  (kg/m3)': 'Limestone Powder (kg/m³)',
    'Quartzpowder (kg/m3)': 'Quartz Powder (kg/m³)',
    'Glass powder (kg/m3)': 'Glass Powder (kg/m³)',
    'Rice husk ash (kg/m3)': 'Rice Husk Ash (kg/m³)',
    'Metakaolin (kg/m³)': 'Metakaolin (kg/m³)',
    'GGBFS  (kg/m³)': 'GGBFS (kg/m³)',
    'Slag Amount   (kg/m³)': 'Slag (kg/m³)',
    'Nano-CaCO3 (kg/m3)': 'Nano-CaCO₃ (kg/m³)',
    'Nano-Al2O3 (kg/m3)': 'Nano-Al₂O₃ (kg/m³)',
    'Nano-TiO2 (kg/m3)': 'Nano-TiO₂ (kg/m³)',
    'Nano Silica (kg/m3)': 'Nano Silica (kg/m³)',
    'Filler (kg/m³)': 'Filler (kg/m³)',
    'Amount  (kg/m³)': 'Fine Aggregate (kg/m³)',
    'Amount / Quantity of Fiber': 'Steel Fibre (kg/m³)',
    'Amount  (kg/m³).2': 'Superplasticizer (kg/m³)',
    '28-Day': '28-Day Compressive Strength (MPa)'
})
```

```python
table_3_1 = (
    report_df
    .describe()
    .T[['min', 'max', 'mean', 'std']]
    .round(2)
)

table_3_1.columns = [
    'Minimum',
    'Maximum',
    'Mean',
    'Standard Deviation'
]

table_3_1
```

```python
table_3_1.to_excel(
    "Tables/Table_3_1_Descriptive_Statistics.xlsx"
)

print("✅ Table 3.1 saved successfully!")
```

```python
# Missing Values Summary

missing = report_df.isnull().sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))

missing.plot(kind='bar')

plt.title("Missing Values per Variable")
plt.xlabel("Variables")
plt.ylabel("Number of Missing Values")

plt.ylim(0,1)      # Makes zero values visible
plt.grid(axis='y', alpha=0.3)

plt.xticks(rotation=75, ha='right')

plt.tight_layout()

plt.savefig(
    "Figures/Figure_3_2_Missing_Values.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.hist(
    report_df["28-Day Compressive Strength (MPa)"],
    bins=25,
    edgecolor="black"
)

plt.title("Distribution of 28-Day Compressive Strength")
plt.xlabel("Compressive Strength (MPa)")
plt.ylabel("Frequency")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig(
    "Figures/Figure_3_2_Compressive_Strength_Distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import seaborn as sns
```

```python
# Correlation Heatmap

plt.figure(figsize=(14,10))

correlation_matrix = report_df.corr(numeric_only=True)

sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Heatmap of the Cleaned UHPC Dataset")

plt.tight_layout()

plt.savefig(
    "Figures/Figure_3_3_Correlation_Heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# Figure 3.4 - Distribution of Key Input Variables

key_variables = [
    "Cement (kg/m³)",
    "Silica Fume (kg/m³)",
    "Fly Ash (kg/m³)",
    "GGBFS (kg/m³)",
    "Fine Aggregate (kg/m³)",
    "Steel Fibre (kg/m³)",
    "Superplasticizer (kg/m³)"
]

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(key_variables):
    axes[i].hist(report_df[col], bins=20, edgecolor="black")
    axes[i].set_title(col, fontsize=10)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Frequency")

# Hide unused subplots
for j in range(len(key_variables), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()

plt.savefig(
    "Figures/Figure_3_4_Input_Variable_Distributions.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
